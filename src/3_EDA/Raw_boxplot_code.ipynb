```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data into a DataFrame
df = pd.read_csv("data/cohorted_data.csv")
df.fillna(0, inplace=True)  # Account for NaNs in subset

# Create empty lists to store boxplot data
boxplot_data = []
labels = []

for diabetes_group in df['diabetes'].unique():
    diabetes_df = df[df['diabetes'] == diabetes_group]
    diabetes_label = 'With Diabetes' if diabetes_group == 1 else 'Without Diabetes'
    
    # Iterate over unique races
    for race_group in diabetes_df['race_group'].unique():
        race_df = diabetes_df[diabetes_df['race_group'] == race_group]

        # Extract the variable you want to plot
        xvar = race_df['totalgluc_perlos']

        # Calculate the first and third quartiles
        Q1 = xvar.quantile(0.25)
        Q3 = xvar.quantile(0.75)

        # Calculate the interquartile range (IQR)
        IQR = Q3 - Q1

        # Define a threshold for outlier detection (e.g., Q1 - 1.5 * IQR or Q3 + 1.5 * IQR)
        lower_threshold = Q1 - 1.5 * IQR
        upper_threshold = Q3 + 1.5 * IQR

        # Remove outliers
        filtered_data = xvar[(xvar >= lower_threshold) & (xvar <= upper_threshold)]

        # Append data and labels to the lists
        combined_label = f'{race_group}\n({diabetes_label})'
        boxplot_data.append(filtered_data)
        labels.append(combined_label)  # Combined label with race_group and diabetes_label

# Create a boxplot with positions
plt.boxplot(boxplot_data, positions=range(1, len(labels) + 1), labels=labels)

# Rotate the labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=10)  # Rotate and adjust font size

# Set labels and title
plt.xlabel('Race and Diabetes Status')
plt.ylabel('Number of glucose measurements\n per LOS')
plt.title('Glucose Measurements per LOS Boxplot Comparison')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
